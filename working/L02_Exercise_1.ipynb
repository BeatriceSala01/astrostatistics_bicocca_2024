**SIMULATE THE MONTY HALL PROBLEM**

```python
import numpy as np
import random
```

```python
doors = ["goat", "goat", "car"]

random.shuffle(doors)

j = 0
index_car = 0

while j<=2:
    if doors[j] == 'car':
        index_car = j
    j = j+1

print("Car's index in the array: ", index_car)
print(doors)
```

```python
Nc = Ns = Nn = 0
N = 5000

i = 0

def choicef(array):
        index_choice = np.random.choice(len(array))
        enter = array[index_choice]
        return index_choice, enter

while i<N:

    index_c, choice_c = choicef(doors) #choice and corresponding index (in the array) for the conservative
    
    #let's find the index that the announcer has to eliminate
    index_p, choice_p = choicef(doors)

    if index_c == index_car:
        while index_p == index_c:
            index_p, choice_p = choicef(doors)
    if index_c != index_car:
        while index_p == index_car or index_p == index_c:
            index_p, choice_p = choicef(doors)
    
    #let's find the index that the switcher chooses (nor the one of the announcer nor the one chosen before)
    index_s, choice_s = choicef(doors)

    while index_s == index_c or index_s == index_p:
        index_s, choice_s = choicef(doors)
    
    #let's find the index that the newcomer chooses (it can be every except for the announcer's one)
    index_n, choice_n = choicef(doors)

    while index_n == index_p:
        index_n, choice_n = choicef(doors)
   
    #let's count how many times the three competitors have found the car
    if index_c == index_car: 
        Nc = Nc + 1
    if index_s == index_car: 
        Ns = Ns + 1
    if index_n == index_car: 
        Nn = Nn + 1
    
    i=i+1
```

```python
print("Conservative: ", Nc, "Switcher: ", Ns, "Newcomer: ", Nn)
```

The probabilities are:

```python
print("Probability for the conservative: ", Nc/N)
print("Probability for the switcher: ", Ns/N)
print("Probability for the newcomer: ", Nn/N)
```

```python
M = 100   #number of the doors
Doors = ["car"]

k = 1

while k<M:
    Doors.append('goat')
    k = k+1

random.shuffle(Doors)

j = 0
index_Car = 0

while j<M:
    if Doors[j] == 'car':
        index_Car = j
    j = j+1

print("Car's index in the array: ", index_Car)
#print(Doors)
```

```python
NC = NS = NN = 0
m = M-2
l = 0

indexes = []
newDoors = Doors.copy()

while l<N:

    index_C, choice_C = choicef(Doors) #choice and corresponding index (in the array) for the conservative

    #let's find the indexes that the announcer has to eliminate
    
    a = 0
    #in the following, I always want only 2 doors, the one chosen by the conserver and another one
    #(if the conserver has found the car, then I want the other one to be an arbitrary goat,
    #if he has found a goat, then I want the car)
    while a<m:  
        index_P, choice_P = choicef(Doors)
        if index_C == index_Car:
            while index_P == index_C:
                index_P, choice_P = choicef(Doors)
        else:
            while index_P == index_Car or index_P == index_C:
                index_P, choice_P = choicef(Doors)
        
        if index_P not in indexes:
            indexes.append(index_P)
            newDoors[index_P] = 0
            a = a + 1
        else:
            a = a
    
    #let's find the newcomer and switcher's doors
    index_S, choice_S = choicef(newDoors)
    index_N, choice_N = choicef(newDoors)
    
    while choice_S == 0 or choice_S == choice_C:
        index_S, choice_S = choicef(newDoors)
        
    while choice_N == 0:
        index_N, choice_N = choicef(newDoors)

    #let's count how many times the three competitors have found the car
    if index_C == index_Car: 
        NC = NC + 1
    if index_S == index_Car: 
        NS = NS + 1
    if index_N == index_Car: 
        NN = NN + 1
    
    indexes.clear()
    newDoors = Doors.copy()
    
    l=l+1
```

```python
print("Conservative: ", NC, "Switcher: ", NS, "Newcomer: ", NN)
```

The probabilities are:

```python
print("Probability for the conservative: ", NC/N)
print("Probability for the switcher: ", NS/N)
print("Probability for the newcomer: ", NN/N)
```
