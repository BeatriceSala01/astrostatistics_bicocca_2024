```python
import numpy as np
import pylab as plt
import scipy.stats
```

```python
deaths = np.array([0,1,2,3,4])
Ncorps = np.array([109,65,22,3,1])
tot = np.sum(Ncorps)

freqDeaths = []
freqDeaths = Ncorps/tot
plt.scatter(deaths, freqDeaths)

mu = 122/200

for Mu in [0.1, 0.3, mu, 0.9, 1.4, 2]:
    p = scipy.stats.poisson(Mu).pmf(deaths)
    plt.plot(deaths, p, label='$\mu='+str(Mu)+"$")

plt.legend()
plt.show()
```

```python
goodmu = np.average(deaths, weights=freqDeaths)
plt.scatter(deaths, freqDeaths)

print(goodmu, mu)

p = scipy.stats.poisson(goodmu).pmf(deaths)
plt.plot(deaths, p)
```
