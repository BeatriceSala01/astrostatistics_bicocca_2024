**PDF OF THE LOG**

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from math import log
```

```python
N = 10000

x = np.random.uniform(0.1, 10, N)
```

```python
fig, graph1 = plt.subplots()
graph1.hist(x, color='C0', bins=100, lw=2,  histtype='step', label='Histogram')  # histtype='step' doesn't fill the isto
```

```python
xi = np.linspace(0.1, 10, N)
yi = np.log10(xi)
y = np.log10(x)

fig, graph2 = plt.subplots()
graph2.hist(y, color='C0', bins=100, lw=2,  histtype='step', label='Histogram', density=True)
ploty = xi*np.log(10)*(1/9.9)
plt.plot(yi, ploty, color='red', label='pdf(x)')
plt.legend()
plt.show()
```

```python
meanx = np.mean(x)
lmeanx = log(meanx, 10)

meany = np.mean(y)

print("Mean x: ", lmeanx)
print("Mean y: ", meany)
```

```python
medianx = np.median(x)
lmedianx = log(medianx, 10)

mediany = np.median(y)

print("Log Median x: ", lmedianx)
print("Log Median y: ", mediany)
```
