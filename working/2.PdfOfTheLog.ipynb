1. Use numpy to draw N (where N is large...) random samples from a uniform distribution between 0.1 and 10, and store these samples as x.
2. Use matplotlib to make a histogram of these samples.
3. Compute the base-10 log of your array x, and store this as y.
4. Make another histogram for y. Using the equation to transform probability distributions, write what the theoretical pdf of y is, and overplot it onto your histogram.
5. Compute the log of the mean of x and the mean of y. Now compute the log of the median of x and the median of y.

You should note that the means are different, but the medians (as it is a cumulative statistic) are the same. The mean is affected by the scale of the sample values, but the median only depends on the ordering of the samples. Monotonic transformations (like taking the log) do not change the ordering of samples.

```python
import numpy as np
from matplotlib import pyplot as plt
```

I define $N=50000$ as the number of data points in the sample, and I draw $N$ random points from a uniform distribution.

```python
N = 50000
x = np.random.uniform(0.1, 10, N)

plt.hist(x, bins=int(N**0.5), edgecolor='dimgrey', histtype='step', density=True, label='Histogram')
plt.legend()
plt.show()
```

I define y as the logarithm of x and I compute the transform probability distribution for the y points as:

\begin{equation}
f(x)=log_{10}(x)
\end{equation}

\begin{equation}
pdf(y)= |dx/dy|pdf(x)=|(10^y)ln(10)|pdf(x)
\end{equation}

```python
y = np.log10(x)

plt.hist(y, bins=int(N**0.5), edgecolor='dimgrey', histtype='step', density=True, label='Histogram')

p_x = 1/(10-0.1)  # this is pdf(x)
Y = y.copy()
Y.sort()
p_y = p_x*(10**Y)*np.log(10)  # this is pdf(y)

plt.plot(Y, p_y, color='orange', label='Fit')
plt.legend()
plt.show()
```

I compare the means and the medians.

```python
meanX = np.mean(x)
log_meanX = np.log10(meanX)
meanY = np.mean(y)
print('Logarithm of the mean: ', log_meanX, '\nMean of the logarithm: ', meanY, '\n')

medianX = np.median(x)
log_medianX = np.log10(medianX)
medianY = np.median(y)
print('Logarithm of the median: ', log_medianX,'\nMedian of the logarithm: ',  medianY)
```

The logarithm of the mean of x and the mean of the logarithm of x (so the mean of y) are different. While the logarithm of the median of x and the median of the logarithm of x (so the median of y) are almost the same.
